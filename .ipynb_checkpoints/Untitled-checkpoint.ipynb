```python
import arff
import pandas as pd

# Charger ton fichier .arff
data = arff.load(open("Autism-Adult-Data.arff"))

# Convertir en DataFrame Pandas
df = pd.DataFrame(data['data'], columns=[attr[0] for attr in data['attributes']])

# Afficher les 5 premières lignes
print(df.head())
print(df.info())
```
